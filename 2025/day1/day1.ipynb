```python
with open('input.txt') as f:
    lines = f.read().splitlines()
```

```python
def click(current: int , action: str):

    incr = int(action[1:])
    direction = 1 if action[0] == 'R' else -1

    zero = 0

    for _ in range(incr):
        current+=direction

        if current == 100:
            current = 0
        elif current == -1:
            current = 99

        if current == 0:
            zero+=1

    return current, zero
```

```python
dial = 50
dial_history = []
for action in lines:
    dial, delta = click(dial, action)
    dial_history.append((dial, delta))
```

```python
# Part 1
sum([d == 0 for d, _ in dial_history])
```

```python
# Part 2
sum([d for _ , d in dial_history])
```
