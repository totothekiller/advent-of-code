```python
data = []
with open('input.txt') as f:
    for line in f.read().splitlines():
        data.append(list(line))
```

```python
x_len, y_len = len(data), len(data[0])
print(x_len, y_len)
```

```python
from collections.abc import Iterator

def around(x,y)-> Iterator[tuple[int, int]]:

    x_min, x_max = max(0, x - 1), min(x + 1, x_len - 1)
    y_min, y_max = max(0, y - 1), min(y + 1, y_len - 1)

    for i in range(x_min, x_max + 1):
        for j in range(y_min, y_max + 1):

            if i == x and j == y:
                continue

            yield (i , j)
```

```python
total = 0

for i in range(0, x_len):
    for j in range(0, y_len):
        if data[i][j] == '@':
            if sum([data[x][y] == '@' for x,y in around(i,j)]) < 4:
                total +=1

#Part 1
print(total) 
```

```python
import copy

def remove(locations) -> int:
    total = 0
    next_data = copy.deepcopy(locations)

    for i in range(0, x_len):
        for j in range(0, y_len):
            if locations[i][j] == '@':
                if sum([locations[x][y] == '@' for x,y in around(i,j)]) < 4:
                    total +=1
                    next_data[i][j]='x'


    return next_data, total
```

```python
current = copy.deepcopy(data)
total = 0

while True:
    current, step = remove(current)
    total+=step
    if step <= 0:
        break

# Part2
print(total)
```
