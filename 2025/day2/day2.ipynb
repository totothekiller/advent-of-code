```python
import re
```

```python
regex = re.compile(r"\b([0-9]+)\1\b")
```

```python
with open('input.txt') as f:
    ranges = f.read().split(",")
```

```python
part1 = 0

for r in ranges:
    start, end = r.split("-")
    for i in range(int(start), int(end) + 1):
        if regex.match(str(i)) is not None:
            part1+=i
```

```python
# Part 1
print(part1)
```

```python
regex2 = re.compile(r"\b([0-9]+)\1+\b")
```

```python
part2 = 0

for r in ranges:
    start, end = r.split("-")
    for i in range(int(start), int(end) + 1):
        if regex2.match(str(i)) is not None:
            part2+=i
```

```python
# Part 2
print(part2)
```
