```python
with open('input.txt') as f:
    lines = f.read().splitlines()
```

```python
def find_joltage(banks: str, joltage_power: int) -> int:

    if joltage_power < 0:
        return 0

    joltage = 0
    idx = -1

    search_range = len(banks) - joltage_power

    # find best battery
    for i in range(9, 0, -1):
        idx = banks[:search_range].find(str(i))
        if  idx > -1:
            joltage = i
            break

    #print(f"{banks} / {joltage_power} / {idx} / {joltage}")

    return pow(10, joltage_power) * joltage + find_joltage(banks[idx+1:], joltage_power-1)
```

```python
# Part 1 
sum(map(lambda b: find_joltage(b, 1), lines))
```

```python
# Part 2
sum(map(lambda b: find_joltage(b, 11), lines))
```
